# Regresi Polinomial Method

```python
import numpy as np
import matplotlib.pyplot as plt

# Nomor poin
n = 200
NPM = 52
```

```python


# Menghasilkan nilai x acak antara 0 hingga 4
np.random.seed(NPM)
x = np.random.uniform(0, 4*np.pi, n)

# Fungsi asli (tanpa noise)
y_true = 4 * np.sin(2*x + 0.2)

# Menambahkan noise Gaussian pada data
noise = np.random.normal(0, 0.4, n)  # rata-rata=0, standar deviasi=0.4
y_noisy = y_true + noise

# Dictionary untuk menyimpan koefisien regresi polinomial
all_coefficients = {}

# Membuat kanvas untuk menampilkan grafik
plt.figure(figsize=(12, 8))

# Menampilkan data noisy dan fungsi aslinya
plt.scatter(x, y_noisy, label="Noisy data", color="red", alpha=0.6)
x_true_plot = np.linspace(0, 4 * np.pi, 500)
plt.plot(x_true_plot, 4 * np.sin(2 * x_true_plot + 0.2), label="True function", color="blue", linewidth=2)

# Melakukan regresi polinomial dari orde 1 hingga 6
for order in range(1, 7):
    # Menghitung koefisien regresi polinomial
    coefficients = np.polyfit(x, y_noisy, order)
    all_coefficients[order] = coefficients

    # Membentuk fungsi polinomial berdasarkan koefisien
    poly_function = np.poly1d(coefficients)

    # Menghasilkan titik-titik untuk kurva hasil fitting
    x_plot = np.linspace(0, 4*np.pi, 500)
    y_fitted = poly_function(x_plot)

    # Menampilkan kurva fitting polinomial
    plt.plot(x_plot, y_fitted, label=f'Polinomial Orde {order}', linestyle='--')

# Menambahkan judul, label sumbu, dan legenda pada grafik
plt.title('Regresi Polinomial dari Orde 1 hingga 6 pada Data Sinusoidal')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.savefig('polynomial_regression_orders_1_to_5.png')

# Menampilkan koefisien untuk setiap orde polinomial
for order, coeffs in all_coefficients.items():
    print(f"\nKoefisien polinomial orde {order}:")
    print(coeffs)
```

Hasil analisis regresi polinomial dari orde 1 hingga 6 menunjukkan bahwa orde rendah (12) cenderung underfitting karena gagal menangkap pola osilasi sinusoidal, sementara orde menengah seperti orde 4 terlihat paling seimbang dalam mengikuti bentuk gelombang tanpa terlalu terjebak pada noise. Namun, ketika orde dinaikkan hingga 5 ke atas, model justru menunjukkan gejala overfitting dengan kurva yang terlalu berayun mengikuti data acak. Dari sisi evaluasi, perhitungan MAPE untuk orde 16 ternyata masih jauh di atas 20%, terutama karena karakteristik fungsi sinusoidal yang memiliki titik mendekati nol sehingga membuat error persentase meledak. Menanggapi saran Bapak agar dicoba fitting yang lebih tinggi hingga mencapai MAPE di bawah 20, akan saya coba lanjutkan dengan menaikkan orde polinomial lebih tinggi lagi, sekaligus mempertimbangkan metrik alternatif seperti RMSE atau MAE yang lebih stabil untuk data sinusoidal, agar hasil analisis bisa lebih akurat dan tetap menjaga kemampuan generalisasi model.

Garis regresi dengan orde tinggi (811) mengikuti titik data noisy (warna merah) secara berlebihan sehingga melenceng jauh dari fungsi aslinya (garis biru). Hal ini tampak dari bentuk kurva yang berliku-liku tidak wajar, terutama di ujung kiri dan kanan grafik, di mana model berusaha menyesuaikan diri dengan noise alih-alih pola utama sinusoidal. Dengan kata lain, meskipun orde tinggi dapat menurunkan error pada data latih, hasilnya justru kehilangan generalisasi karena lebih banyak menghafal noise daripada merepresentasikan fungsi sebenarnya.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error

# --- Data ---
np.random.seed(0)
x = np.linspace(0, 12, 100)
y_true = 4 * np.sin(x)  # true function
noise = np.random.normal(0, 0.5, size=x.shape)
y_noisy = y_true + noise

# --- Hitung regresi polinomial orde 1-7 ---
MAPE_scores = {}

for degree in range(1, 8):
    # fitting polinomial
    coeffs = np.polyfit(x, y_noisy, degree)
    poly = np.poly1d(coeffs)
    
    # prediksi
    y_pred = poly(x)
    
    # hitung MAPE terhadap true function
    mape = mean_absolute_percentage_error(y_true, y_pred) * 100
    MAPE_scores[f"Orde {degree}"] = mape
    
    # print hasil
    print(f"MAPE Orde {degree}: {mape:.2f}%")

```

Dari hasil perhitungan yang saya lakukan, nilai MAPE yang muncul justru sangat besar, contohnya pada Orde 1 sekitar 9.23 x 10^15%, Orde 2 sekitar 1.00 x 10^16%, Orde 3 sekitar 2.46 x 10^15%, Orde 4 sekitar 1.67 x 10^17%, Orde 5 sekitar 8.89 x 10^15%, Orde 6 sekitar 3.63 x 10^15%, dan Orde 7 sekitar 5.73 x 10^15%. Padahal sesuai dengan revisi dari Bapak, saya diarahkan untuk menggunakan regresi polinomial dengan orde yang lebih tinggi agar diperoleh nilai MAPE di bawah 20% sebagai indikator akurasi model. Oleh karena itu, saya mencoba menggunakan orde 6 karena menurut pengamatan tidak terjadi overfitting. Namun, hasil yang muncul tetap tidak masuk akal, sebab normalnya nilai MAPE berada di kisaran beberapa persendi bawah 10% dikategorikan sangat baik, dan di bawah 20% masih cukup dapat diterima. Besarnya nilai MAPE ini kemungkinan terjadi karena adanya data aktual yang mendekati nol sehingga menyebabkan error persentase membengkak, atau terdapat kesalahan dalam implementasi rumus MAPE.
