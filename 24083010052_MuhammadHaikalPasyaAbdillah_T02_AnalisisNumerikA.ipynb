### 2. Integral dari fungsi f untuk interval x

```python
F_total = np.cumsum((f[:-1] + f[1:]) / 2 * h)   
F_total = np.insert(F_total, 0, 0)

print("Nilai integral total   :", F_total[-1])

plt.figure(figsize=(8,4))
plt.plot(x, F_total, 'k', label="Integral total")
plt.title("Integral f(x)")
plt.legend(); plt.grid(True)
plt.show()
```

Hasil dari integral total adalah 2.19 jika dibulatkan. Oleh karena itu, luas yang berada di atas sumbuh x lebih dari luas yang ada di bawah sumbu x, jadi nilai integralnya adalah positif.



### 3. Perhitungan secara terpisah bagian integral di atas dan di bawah y=0

```python
# Memisahkan bagian atas dan bawah
f_upper = np.clip(f, 0, None)
f_lower = np.clip(f, None, 0)

# Menghitung integral dengan metode trapezoidal (luas Lingkaran)
area_upper = np.trapz(f_upper, x)
area_lower = np.trapz(f_lower, x)

print("Luas integral di atas sumbu x  :", area_upper)
print("Luas integral di bawah sumbu x :", area_lower)

plt.figure(figsize=(8,5))
plt.plot(x, f, 'k', label='f(x)')
plt.fill_between(x, f_upper, 0, color='green', alpha=0.5, label='Area atas')
plt.fill_between(x, f_lower, 0, color='red', alpha=0.5, label='Area bawah')
plt.axhline(0, color='black', linewidth=0.8)
plt.title("Integral f(x) di atas & bawah sumbu x")
plt.legend()
plt.grid(True)
plt.show()
```

Grafik menunjukkan bahwa area hijau di atas sumbu x lebih dominan daripada area merah di bawahnya. Artinya, fungsi ini lebih banyak memberi “sumbangan positif” daripada negatif, sehingga hasil integral total tetap bernilai positif.
