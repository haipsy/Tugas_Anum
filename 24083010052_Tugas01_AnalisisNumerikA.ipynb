```python
# nama: Mauhammad Haikal Pasya Abdillah
# NPM: 24081020052

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# range harga jual (satuan: ribuan)
x = np.linspace(3, 10, 500)  # dari 3 (3000) sampai 10 (10000)

# distribusi penjualan (bentuk skewed, supaya mirip contoh)
a = 6
loc = 3.5
scale = 1.5
y = skewnorm.pdf(x, a=a, loc=loc, scale=scale) * 10

# seed sesuai tugas
np.random.seed(520)
y = y + np.random.normal(0, 0.005, size=x.shape)  # noise kecil
y = y * 1000  # skala jumlah penjualan

# -----------------------------
# harga produksi (garis merah) turun dari kiri ke kanan,
# ujung kanan tepat pada x=10 karena kita pakai array x yang sama.
harga_produksi = np.linspace(3.6, 2.6, len(x))  # <-- ubah angka 3.6 (awal) dan 2.6 (akhir) di sini
# -----------------------------

# hitung laba: (harga jual - harga produksi) * jumlah terjual (dalam ribuan)
laba = (x - harga_produksi) * (y / 1000)

# titik optimal (harga jual yang memaksimalkan laba)
idx_opt = np.argmax(laba)
x_opt = x[idx_opt]
laba_opt = laba[idx_opt]

# Plot
plt.figure(figsize=(11,6))
plt.plot(x, y/1000, color="navy", linewidth=2, label="Penjualan")       # biru
plt.plot(x, harga_produksi, color="red", linewidth=2, label="Harga Produksi")  # merah turun
plt.plot(x, laba, color="green", linewidth=1.5, label="Laba")           # hijau
plt.axvline(x_opt, color="brown", linestyle="--", linewidth=2, label=f"Harga Optimal = {x_opt:.2f}")

plt.title("Analisis Laba Toko Roti Maknyus")
plt.xlabel("x (Harga Jual)")
plt.ylabel("Jumlah barang yang dibuat/laku")
plt.legend(loc="upper right")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()

print(f"Harga optimal penjualan roti: {x_opt:.2f}")
print(f"Keuntungan maksimum: {laba_opt:.2f}")
```

Berdasarkan hasil perhitungan, volume produksi yang memberikan keuntungan maksimum terjadi pada harga sekitar Rp 5,02 per roti dengan total keuntungan sekitar 5,46 satuan keuntungan. Hal ini terjadi karena ketika harga terlalu rendah, memang jumlah roti yang terjual lebih banyak, tetapi keuntungan tidak maksimal sebab biaya produksi justru lebih besar dibanding pendapatan. Sebaliknya, jika harga ditetapkan terlalu tinggi, jumlah pembeli berkurang drastis sehingga total keuntungan juga ikut menurun. Oleh karena itu, harga Rp 5,02 dapat dianggap sebagai titik sweet spot yang seimbang, di mana penjual masih bisa menarik minat pembeli dan pada saat yang sama memperoleh keuntungan paling optimal.
